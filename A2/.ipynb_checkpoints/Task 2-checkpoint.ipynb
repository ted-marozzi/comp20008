```python
import pandas as pd
from sklearn import preprocessing


yeast = pd.read_csv("all_yeast.csv")
yeast.head()

for col_name in yeast:
    col = yeast[col_name]

    if col.isnull().values.any() == True:
        yeast[col_name] = col.fillna(col.median())
        print(col_name)
        print("Min:", col.min())
        print("Median:", col.median())
        print("Max:", col.max())
        print("Mean:", col.mean())
        print("Std:", col.std())
        print("")
    
        
print("""The median imputation method is the most suitable method as
it is not unaffected by skewed data as much as the mean imputation method.""")

yeast_num = yeast[['mcg','gvh','alm','mit','erl','pox','vac','nuc']]


yeast_num = yeast_num.subtract(yeast_num.mean())

print("\nMean centering\n")

for col_name in yeast_num:
    col = yeast_num[col_name]
    print(col_name)
    print("Min:", col.min())
    print("Median:", col.median())
    print("Max:", col.max())
    print("Mean:", col.mean())
    print("Std:", col.std())
    print("")
    
scaler = preprocessing.StandardScaler()

scaled_df = scaler.fit_transform(yeast_num)
scaled_df = pd.DataFrame(scaled_df, columns=yeast_num.columns)
```
